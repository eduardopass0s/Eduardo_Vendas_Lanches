- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 2
class Lanche:
    def __init__(self, nome, ingredientes, preco, tipo_molho=None):
        self.nome = nome
        self.ingredientes = ingredientes
        self.preco = preco
        self.tipo_molho = tipo_molho

    def exibir_dados(self):
        print(f"Lanche: {self.nome} | Ingredientes: {', '.join(self.ingredientes)} | Pre√ßo: R${self.preco:.2f}", end="")
        if self.tipo_molho:
            print(f" | Molho: {self.tipo_molho}")
        else:
            print()

    def aplicar_desconto(self, percentual):
        self.preco -= self.preco * (percentual / 100)
        print(f"Novo pre√ßo com {percentual}% de desconto: R${self.preco:.2f}")

    def verificar_categoria_preco(self):
        categoria = "caro" if self.preco >= 20 else "acess√≠vel"
        print(f"O lanche '{self.nome}' √© {categoria}.")

class LancheVegano(Lanche):
    def __init__(self, nome, ingredientes, preco, tipo_molho=None, tipo_proteina="Vegetal"):
        super().__init__(nome, ingredientes, preco, tipo_molho)
        self.tipo_proteina = tipo_proteina

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Tipo de prote√≠na: {self.tipo_proteina}")

    def alterar_tipo_proteina(self, novo_tipo):
        self.tipo_proteina = novo_tipo
        print(f"Tipo de prote√≠na alterado para: {self.tipo_proteina}")

lanche1 = Lanche("X-Burger", ["P√£o", "Carne", "Queijo"], 12.50)
lanche2 = LancheVegano("X-Vegano", ["P√£o", "Alface", "Tomate"], 18.00, "Molho Pesto", "Tofu")

lanche1.exibir_dados()
lanche2.exibir_dados()

lanche1.aplicar_desconto(10)
lanche2.aplicar_desconto(5)

lanche1.verificar_categoria_preco()
lanche2.verificar_categoria_preco()

lanche2.alterar_tipo_proteina("Seitan")
lanche2.exibir_dados()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**

- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 3
class Lanche:
    def __init__(self, nome, ingredientes, preco, tipo_molho=None):
        self.nome = nome
        self.ingredientes = ingredientes
        self.__preco = preco 
        self.tipo_molho = tipo_molho

    def get_preco(self):
        return self.__preco

    def set_preco(self, novo_preco):
        if novo_preco >= 0:
            self.__preco = novo_preco
            print(f"Pre√ßo atualizado para: R${self.__preco:.2f}")
        else:
            print("Erro: O pre√ßo n√£o pode ser negativo.")

    def aplicar_desconto(self, percentual):
        desconto = self.__preco * (percentual / 100)
        self.__preco -= desconto
        print(f"Novo pre√ßo com {percentual}% de desconto: R${self.__preco:.2f}")

    def calcular_preco_final(self, percentual_imposto):
        preco_com_imposto = self.__preco * (1 + percentual_imposto / 100)
        return preco_com_imposto

    def exibir_dados(self):
        print(f"Lanche: {self.nome} | Ingredientes: {', '.join(self.ingredientes)} | Pre√ßo: R${self.__preco:.2f}", end="")
        if self.tipo_molho:
            print(f" | Molho: {self.tipo_molho}")
        else:
            print()

lanche1 = Lanche("X-Burger", ["P√£o", "Carne", "Queijo"], 12.50)
lanche2 = Lanche("X-Salada", ["P√£o", "Carne", "Queijo", "Alface"], 14.00, "Maionese")

lanche1.exibir_dados()
lanche2.exibir_dados()

lanche1.aplicar_desconto(10)
lanche2.aplicar_desconto(5)

preco_final_lanche1 = lanche1.calcular_preco_final(10)
preco_final_lanche2 = lanche2.calcular_preco_final(10)

print(f"Pre√ßo final do X-Burger com imposto: R${preco_final_lanche1:.2f}")
print(f"Pre√ßo final do X-Salada com imposto: R${preco_final_lanche2:.2f}")

lanche1.set_preco(15.00)  
lanche1.exibir_dados() 
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 4

class Lanche:
    def __init__(self, nome, ingredientes, preco, tipo_molho=None):
        self.nome = nome
        self.ingredientes = ingredientes
        self.__preco = preco  # Atributo privado
        self.tipo_molho = tipo_molho

    def get_preco(self):
        return self.__preco

    def set_preco(self, novo_preco):
        if novo_preco >= 0:
            self.__preco = novo_preco
            print(f"Pre√ßo atualizado para: R${self.__preco:.2f}")
        else:
            print("Erro: O pre√ßo n√£o pode ser negativo.")

    def aplicar_desconto(self, percentual):
        self.__preco -= self.__preco * (percentual / 100)
        print(f"Novo pre√ßo com {percentual}% de desconto: R${self.__preco:.2f}")

    def calcular_preco_final(self, percentual_imposto):
        preco_com_imposto = self.__preco * (1 + percentual_imposto / 100)
        return preco_com_imposto

    def exibir_dados(self):
        print(f"Lanche: {self.nome} | Ingredientes: {', '.join(self.ingredientes)} | Pre√ßo: R${self.__preco:.2f}", end="")
        if self.tipo_molho:
            print(f" | Molho: {self.tipo_molho}")
        else:
            print()

# Subclasse (LancheVegano) que herda da classe Lanche
class LancheVegano(Lanche):
    def __init__(self, nome, ingredientes, preco, tipo_molho=None, tipo_proteina="Vegetal"):
        super().__init__(nome, ingredientes, preco, tipo_molho)
        self.tipo_proteina = tipo_proteina

    def exibir_dados(self):
        super().exibir_dados()  # Exibe os dados da classe pai
        print(f"Tipo de prote√≠na: {self.tipo_proteina}")

# Banco de Dados Simulado em Mem√≥ria
class BancoDeDados:
    def __init__(self):
        self.lanches_db = []  # Lista que simula o banco de dados

    def inserir_lanche(self, lanche):
        # Inserir um lanche na "tabela"
        self.lanches_db.append(lanche)
        print(f"Lanche '{lanche.nome}' inserido no banco de dados.")

    def listar_lanches(self):
        # Listar todos os lanches na "tabela"
        if not self.lanches_db:
            print("Nenhum lanche encontrado no banco de dados.")
            return
        print("\n--- Lanches no Banco de Dados ---")
        for lanche in self.lanches_db:
            lanche.exibir_dados()

    def excluir_lanche(self, nome):
        # Excluir um lanche pelo nome
        for lanche in self.lanches_db:
            if lanche.nome == nome:
                self.lanches_db.remove(lanche)
                print(f"Lanche '{nome}' removido do banco de dados.")
                return
        print(f"Lanche '{nome}' n√£o encontrado no banco de dados.")

# Testando o c√≥digo
def main():
    banco = BancoDeDados()

    # Criando objetos de lanches
    lanche1 = Lanche("X-Burger", ["P√£o", "Carne", "Queijo"], 12.50)
    lanche2 = LancheVegano("X-Vegano", ["P√£o", "Alface", "Tomate"], 18.00, "Molho Pesto", "Tofu")

    # Inserindo lanches no banco de dados
    banco.inserir_lanche(lanche1)
    banco.inserir_lanche(lanche2)

    # Listando lanches no banco
    banco.listar_lanches()

    # Excluindo um lanche do banco
    banco.excluir_lanche("X-Burger")  # Excluindo o lanche "X-Burger"

    # Listando lanches ap√≥s a exclus√£o
    print("\n--- Lanches ap√≥s exclus√£o ---")
    banco.listar_lanches()

if __name__ == "__main__":
    main()
```
